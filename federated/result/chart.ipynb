```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
models = json.load(open('models.json', 'r'))
x = models['num_clients']
```

```python
plt.plot(models['lstm'], 'o-r', label='lstm')
plt.plot(models['bidirectional'], 'o-b', label='bidirectional lstm')
plt.plot(models['gru'], 'o-g', label='gru')

xi = list(range(len(x)))
plt.xticks(xi, x)

plt.xlabel('Number of Client')
plt.ylabel('MSE')

plt.legend()
plt.show()
```
